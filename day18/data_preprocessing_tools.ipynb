# Data Preprocessing Tools

## Importing the libraries

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('D:/ML Learnings/Udemy ml a-z/part 1 data/Data.csv')
```

```python
df
```

## Importing the dataset

```python
X = df.iloc[:,:-1].values
X
```

```python
y = df.iloc[:,-1].values
y
```

## Taking care of missing data

```python
from sklearn.impute import SimpleImputer
```

```python
imp = SimpleImputer(missing_values=np.nan,strategy='mean')
imp.fit(X[:,1:3])
X[:,1:3] = imp.transform(X[:,1:3])
```

```python
X
```

## Encoding categorical data

### Encoding the Independent Variable

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
ct = ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[0])] ,remainder='passthrough')
```

```python
X = ct.fit_transform((X))
X
```

### Encoding the Dependent Variable

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
y = le.fit_transform(y)
y
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
print('X_train''\n',X_train,'\n''\n''X_test''\n',X_test,'\n''\n''y_train''\n',y_train,'\n''\n''y_test''\n',y_test) 
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
```

```python
X_train[:,3:5] = ss.fit_transform(X_train[:,3:5])
```

```python
X_train
```

```python
X_test[:,3:5] = ss.transform(X_test[:,3:5])
```
