```python
import numpy as np
```

```python
cal_matrix = np.array([[56,3.3,4.4,68.0],
                       [1.2,3.1,52.0,8.0],
                       [1.8,0.3,99.0,0.9]])

print("The Calorie matrix : %s" % cal_matrix,"\n")

cal_sum = cal_matrix.sum(axis=0)#axis = 0 ---> Python sums vertically
print("The summation of calories for different foods : %s\n" %cal_sum)


percentage_cal = 100*cal_matrix/(cal_sum.reshape(1,4))
print("Reshaped per cal to total cal : %s " % (cal_sum.reshape(1,4)),"\n")#cal_sum is already a 1x4 matrix so technically we don't need to call reshape.

print("The percentage of Calories per Nutrient to total Calories : %s" % percentage_cal )
```
