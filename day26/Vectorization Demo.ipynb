```python
import time 

a= np.random.rand(1000000)
b= np.random.rand(1000000)

tic = time.time()
c = np.dot(a,b)
toc=time.time()

print("The Vectorized version : %s \nComputation time : %s ms\n " % (c,str(1000*(toc-tic))))

c = 0
tic - time.time()
for i in range(100000):
    c+= a[i]*b[i]
toc=time.time()
    
print("The Vectorized version : %s \nComputation time : %s ms " % (c,str(1000*(toc-tic))))    
```
